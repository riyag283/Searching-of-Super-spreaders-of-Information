```python
def common(m,n):
    mset = set(m)
    nset = set(n)
    return (mset & nset)
```

```python
def unique(list1): 
    list_set = set(list1) 
    unique_list = (list(list_set)) 
    return unique_list 
```

```python
def getIndex(listOfElements, element):
    indexPosList = []
    indexPos = 0
    while True:
        try:
            indexPos = listOfElements.index(element, indexPos)
            indexPosList.append(indexPos)
            indexPos += 1
        except ValueError as e:
            break
    return indexPosList
```

```python
ks = df['ks']
ks = unique(ks)
kdeg = df['kdeg']
kdeg = unique(kdeg)
```

```python
l1 = list(df['ks'])
l2 = list(df['kdeg'])
```

```python
tdf = pd.DataFrame(columns = ['X','Y','value'])
```

```python
for i in ks:
    for j in kdeg:
        list1 = getIndex(l1,i)
        list2 = getIndex(l2,j)
        cmn = common(list1,list2)
        n = len(cmn)
        sum1 = 0
        if n > 15:
            for node in cmn:
                sum1 += df['m'][node]
            sum1 = sum1/n
            sum1 = math.log(sum1)
        new_row = {'X':i,'Y':j,'value':sum1}
        tdf = tdf.append(new_row,ignore_index=True)
```

```python
tdf.to_csv('myfile1.csv')
```
