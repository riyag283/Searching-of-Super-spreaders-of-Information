```python
import pandas as pd
import math
df = pd.read_csv('mynodes.csv',index_col=0)
```

```python
def common(m,n):
    mset = set(m)
    nset = set(n)
    return (mset & nset)
```

```python
def unique(list1): 
    list_set = set(list1) 
    unique_list = (list(list_set)) 
    return unique_list 
```

```python
def getIndex(listOfElements, element):
    indexPosList = []
    indexPos = 0
    while True:
        try:
            indexPos = listOfElements.index(element, indexPos)
            indexPosList.append(indexPos)
            indexPos += 1
        except ValueError as e:
            break
    return indexPosList
```

```python
val = []
for i in range(len(df['ksum'])):
    if df['ksum'][i] == 0:
        val.append(0)
        continue
    val.append(round(math.log(df['ksum'][i],1.00001)))
```

```python
df['ksum'] = val
```

```python
math.log(val[3],10)
```

```python
ks = df['ksum']
ks = unique(ks)
kdeg = df['pr']
kdeg = unique(kdeg)
```

```python
l1 = list(df['ksum'])
l2 = list(df['pr'])
```

```python
tdf = pd.DataFrame(columns = ['X','Y','value'])
for i in ks:
    if i > 465400:
        continue
    for j in kdeg:
        if j > 35:
            continue
        list1 = getIndex(l1,i)
        list2 = getIndex(l2,j)
        cmn = common(list1,list2)
        n = len(cmn)
        sum1 = 0
        if n > 15:
            for node in cmn:
                sum1 += df['m'][node]
            sum1 = sum1/n
            sum1 = math.log(sum1)
        if sum1 < 9.5:
            sum1 = 9.5
        new_row = {'X':i,'Y':round(j),'value':sum1}
        tdf = tdf.append(new_row,ignore_index=True)
```
