```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ks-kdeg.csv')

r = pd.DataFrame()
r['kdeg'] = df['Y']
r['ks'] = df['X']
r['value'] = df['value']
table = r.pivot('kdeg','ks','value')
ax = sns.heatmap(table, cmap='Greens')
ax.invert_yaxis()
plt.show()
```
